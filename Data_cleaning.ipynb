```python
import pandas as pd
import numpy as np
df=pd.read_csv('my_file (1).csv')
df.isnull().sum()
df.columns
```

```python
# Forward and Backward Fill 
Forward_Imputation=df[['Peak','All Time Peak']].ffill()
# print(Forward_Imputation)
print(df)
df[['Peak', 'All Time Peak']] = Forward_Imputation
```

```python

print(df.head())
df.isnull().sum()
```

```python
#Mean, Median and Mode Imputation
df['Peak'] = pd.to_numeric(df['Peak'], errors='coerce')
df['All Time Peak'] = pd.to_numeric(df['All Time Peak'], errors='coerce')
Mean_Imputation=df[['Peak','All Time Peak']].fillna(df['Peak'].mean())
print(Mean_Imputation,"\n")
Median_Imputation=df[['Peak','All Time Peak']].fillna(df['Peak'].median())
print("\n",Median_Imputation)
Mode_Imputation=df[['Peak','All Time Peak']].fillna(df[['Peak','All Time Peak']].mode())
```
